```python
############
import cv2
import os
import time
import numpy as np
import random
import glob
import shutil
import yaml
import torch
from ultralytics import YOLO
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
############### q1.2
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def detect_faces(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=8, minSize=(30, 30))
    return faces

frames_dir = 'frames/'

output_dir = 'frames_with_faces/'
os.makedirs(output_dir, exist_ok=True)


frame_files = os.listdir(frames_dir)
time_intervals = []
for frame_file in frame_files:
    frame_path = os.path.join(frames_dir, frame_file)
    frame = cv2.imread(frame_path)
    start_time = time.time()
    faces = detect_faces(frame)
    end_time = time.time()

    time_interval = end_time - start_time
    time_intervals.append(time_interval)

    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

    output_path = os.path.join(output_dir, frame_file)
    cv2.imwrite(output_path, frame)

average_time_interval = sum(time_intervals) / len(time_intervals)
print(f"Average time taken per frame: {average_time_interval} seconds")
```

the algorithm <br>
<ol>
    <li> classifies a lot of non-faces as faces when minNeighbors are set as low</li>
    <li> classifies a lot of non-faces as faces when scale factor is kept low as it percives some non-face elements as a small face</li>
    <li> a face is not detected if its at inverted upwards or tilted at an angle </li>
    <li>it can detect somewhat blurry face if its big enough( with the scale factor and minNeighbors set upside , may detect smaller blurry face too if we reduce scale factor)</li>
    <li>due to video quality being low some faces are not detected</li>
</ol>

```python
################  q1.4
def calculate_iou(box1, box2):
    x1_tl, y1_tl, w1, h1 = box1
    x2_tl, y2_tl, w2, h2 = box2

    x1_br, y1_br = x1_tl + w1, y1_tl + h1
    x2_br, y2_br = x2_tl + w2, y2_tl + h2

    x_overlap = max(0, min(x1_br, x2_br) - max(x1_tl, x2_tl))
    y_overlap = max(0, min(y1_br, y2_br) - max(y1_tl, y2_tl))

    intersection_area = x_overlap * y_overlap
    box1_area = w1 * h1
    box2_area = w2 * h2
    union_area = box1_area + box2_area - intersection_area

    iou = intersection_area / union_area
    return iou

def detect_faces(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5, minSize=(30, 30))
    return faces

frames_dir = 'frames_with_faces/'

output_dir = 'frames_with_faces_ids'
os.makedirs(output_dir, exist_ok=True)

frame_files = sorted(os.listdir(frames_dir))

tracks = []
track_id_counter = 0  # Initialize the track ID counter

frame_count = 0
for filename in frame_files:
        frame = cv2.imread(os.path.join(frames_dir, filename))
        
        faces = detect_faces(frame)
        
        if frame_count > 0:
            for face in faces:
                x, y, w, h = face
                # Calculate the centroid of the detected face
                centroid = np.array([x + w / 2, y + h / 2])

                best_match_id = None
                best_iou = 0
                for track_id, track in tracks:
                    last_box = track[-1][1]  # Last bounding box in the track
                    iou = calculate_iou(last_box, face)
                    if iou > 0.5 and iou > best_iou:
                        best_match_id = track_id
                        best_iou = iou

                if best_match_id is not None:
                    for track_id, track in tracks:
                        if track_id == best_match_id:
                            track.append((frame_count, face))
                            break
                else:
                    track_id_counter += 1
                    tracks.append((track_id_counter, [(frame_count, face)]))
        else:
            for face in faces:
                track_id_counter += 1
                tracks.append((track_id_counter, [(frame_count, face)]))
        
        for track_id, track in tracks:
            for (frame_number, (x, y, w, h)) in track:
                if frame_number == frame_count:
                    cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
                    cv2.putText(frame, str(track_id), (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)
        
        cv2.imwrite(os.path.join(output_dir, filename), frame)
        
        frame_count += 1
       

unique_tracks = len(tracks)
print("Number of unique tracks created in the first 30 seconds:", unique_tracks)
```

```python
def remove_cache_files(directory):
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.cache'):
                file_path = os.path.join(root, file)
                try:
                    os.remove(file_path)
                    print(f"Removed: {file_path}")
                except OSError as e:
                    print(f"Error: {file_path} - {e}")
```

```python

def train_model(model_name,project,name, config_path, epochs,pretrained):
    destination = f"./{name}_{project[-6:]}.pt"

    if not os.path.exists(destination):
        model = YOLO(model_name)
        results = model.train(data=config_path, epochs=epochs, project=project, name=name, exist_ok=True,pretrained=pretrained,workers=8)
        source = f"./{project}/{name}/weights/best.pt"
        shutil.move(source, destination)
        remove_cache_files("./datasets/")
        return results
    else:
        print("already trained")
        return None

variants = [
    ("yolov8n.pt","yolov8_100n","yolov8n", "./datasets/archive1/config.yaml", 10,True),
    ("yolov8n.yaml","yolov8_100n_new","yolov8n_new", "./datasets/archive1/config.yaml", 10,False),
    ("yolov8m.pt","yolov8_100m","yolov8m", "./datasets/archive1/config.yaml", 10 ,True),
    ("yolov8n.yaml","yolov8_400n_new","yolov8n_new", "./datasets/archive/config.yaml", 10,False),
    ("yolov8n.pt","yolov8_400n","yolov8n", "./datasets/archive/config.yaml", 10,True),
    ("yolov8m.pt","yolov8_400m","yolov8m", "./datasets/archive/config.yaml", 10,True),
]

results=[]
for variant_name,project,name, config_path, epochs ,pretrained in variants:
    print(f"Training: {variant_name}")
    results.append(train_model(variant_name,project,name, config_path, epochs ,pretrained))
```

```python
def is_weight(filename):
    weights_extensions = [".pt"]
    _, ext = os.path.splitext(filename)
    return ext.lower() in weights_extensions
    
def get_weights_paths(directory):
    image_paths = []
    for file in os.listdir(directory):
        if os.path.isfile(os.path.join(directory, file)):
            if is_weight(file):
                image_paths.append(os.path.join(directory, file))    
    return image_paths

weight_paths = get_weights_paths("./")
weight_paths.sort()
print(weight_paths)

```

```python
results = {}

for weight in weight_paths:
    model = YOLO(weight)
    
    temp = "archive1" if "100" in weight else "archive"

    remove_cache_files("./datasets/")
    val_map50 = model.val(data=f'./datasets/{temp}/config.yaml',workers=8).box.map50
    remove_cache_files("./datasets/")
    train_map50 = model.val(data=f'./datasets/{temp}/val_config.yaml',workers=8).box.map50 
    
    results[weight] = {
        'train_map50': str(train_map50),
        'val_map50': str(val_map50)
    }

with open('results.yaml', 'w') as results_file:
    yaml.dump(results, results_file, default_flow_style=False)

print("Results saved to results.yaml file.")
results
```
